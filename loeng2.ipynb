```python
from pprint import pprint
import re

words_count = dict()
with open('/Users/aleksandr.tavgen/Downloads/speeches.txt', 'r') as inputfile:
    for line in inputfile:
        words_list = line.split(' ')
        #print(words_list)
        for word in words_list:
            word = word.lower()
            word = re.sub('[^A-Za-z0-9]+', '', word)
            if word in words_count:
                #words_count[word] = words_count[word] + 1
                words_count[word] += 1
            else:
                words_count[word] = 1

        
ordered_words_count = sorted(words_count, key=words_count.get, reverse=True)
for word in ordered_words_count:
    pprint(word + ' ' + str(words_count[word]))
```
